```python
#query diubah sesuai kebutuhan
query = '(#lowongankerja OR #loker OR "lowongan" OR "lowongan kerja") -is:retweet lang:id'

#komplek query
#query = '(#NewYearsResolution OR #NewYearsResolutions OR #NewYearResolution OR #NewYearResolutions OR #NYResolution OR #NYResolutions OR #NYResolution2022 OR #NYResolutions2022 OR #NewYearsResolution2022 OR #NewYearResolution2022 OR #NewYearsResolutions2022 OR "New Year‚Äôs Resolution" OR "New Years Resolution" OR "New Year‚Äôs Resolutions" OR "New Years Resolutions") lang:id'

response = client.search_recent_tweets(
    query,
    max_results = 100,  #diambil 10 tweet aj, diubah nilai maksimal aj
    tweet_fields = ['author_id','created_at','text','source','lang','geo', 'attachments'],
    user_fields = ['name','username','location','verified'],
    place_fields = ['country','country_code'],
    media_fields = ['duration_ms', 'height', 'media_key', 'preview_image_url', 'public_metrics', 'type', 'url', 'width'],
    expansions = ['geo.place_id', 'author_id', 'attachments.media_keys']
)
```

```python
len(response.data)
```

```python
users = {u["id"]: u for u in response.includes['users']}
media = {m["media_key"]: m for m in response.includes['media']}

for tweet in response.data:
    
    if users[tweet.author_id]:
        print("User ID   : ", tweet.author_id)      # print the author id of the tweet
        user = users[tweet.author_id]
        print("User Name : ", user.username)
        print("Location  : ", user.location)
        print("Verified  : ", user.verified)
    
    print("Datetime  : ", tweet.created_at)
    print("Source    : ", tweet.data['source'])       
    
    print("--- TWEET ---")
    print(tweet.text)
    print()
    print("--- ATTACHMENTS ---")
    if 'attachments' in tweet.data:
        attachments = tweet.data['attachments']
        #print(type(attachments))
        print(attachments)
        media_keys = attachments['media_keys']
        print("Num media: ", len(media_keys))
          
        print("Media link: ")
        for media_key in media_keys:
            print(media[media_key].type)
            if media[media_key].url:
                print(media[media_key].url)
            if media[media_key].preview_image_url:
                print(media[media_key].preview_image_url)
    else:
        print("No attachment")
        
    print("-------")
    print()
    print()
```
